##### Feature Enginering

```python
# changing the Data of Journey to datetime format

df_final['Date_of_Journey'] = pd.to_datetime(df_final['Date_of_Journey'])
df_final.dtypes
```

```python
# Creting Year, month and day column

df_final['Year'] = df_final['Date_of_Journey'].dt.year
df_final['Month'] = df_final['Date_of_Journey'].dt.month
df_final['Date'] = df_final['Date_of_Journey'].dt.day
df_final.head()
```

```python
# can also use the below to get the new features and then change the dataype of the column
df_final["Date"]=df_final['Date_of_Journey'].apply(lambda x:x.split("/")[0])
df_final["Month"]=df_final['Date_of_Journey'].apply(lambda x:x.split("/")[1])
df_final["Year"]=df_final['Date_of_Journey'].apply(lambda x:x.split("/")[2])

# or use the below

cols=['date','month','year']
for i,col in enumerate(cols): 
    df_final[col]=df_final['Date_of_Journey'].apply(lambda x : int(x.split('/')[i]))
```

```python
df_final.info()
```

```python
# Drop the Date of Journey column

df_final.drop('Date_of_Journey',axis=1,inplace=True)
df_final.head()
```

```python
# Working with Arrival time

df_final['Arrival_Time'].str.split(' ').str[0]
```

```python
df_final['Arrival_Time'] = df_final['Arrival_Time'].apply(lambda x : x.split(' ')[0])
df_final.head()
```

```python
df_final.isnull().sum()
```

```python
# Getting arrival hour and min from arrival time

cols  = ['Arrival_hour','Arrival_min']
for i,col in enumerate(cols):
    df_final[col] =  df_final['Arrival_Time'].apply(lambda x : x.split(":")[i])
```

```python
df_final.head()
```

```python
df_final.dtypes
```

```python
# Changing the datatype of arrival hour and arrival min

df_final['Arrival_hour'] = df_final['Arrival_hour'].astype(int)
df_final['Arrival_min'] = df_final['Arrival_min'].astype(int)

df_final.dtypes
```

```python
# drop the arrival time column

df_final.drop('Arrival_Time',axis=1,inplace=True)
df_final.head()
```

```python
# Getting Dept hour and min from Dept time

cols  = ['Dept_hour','Dept_min']
for i,col in enumerate(cols):
    df_final[col] =  df_final['Dep_Time'].apply(lambda x : x.split(":")[i])
```

```python
df_final.head(1)
```

```python
# Changing the datatype of Dept hour and Dept min

df_final['Dept_hour'] = df_final['Arrival_hour'].astype(int)
df_final['Dept_min'] = df_final['Arrival_min'].astype(int)

df_final.dtypes
```

```python
# drop the dep time column

df_final.drop('Dep_Time',axis=1,inplace=True)
df_final.head(2)
```

```python
df_final.info()
```

```python
# Working with route and number of stops
df_final['Total_Stops'].unique()
```

```python
# mapping the total number of stops
df_final['Total_Stops']=df_final['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4})
df_final['Total_Stops'].unique()
```

```python
df_final[df_final['Total_Stops'].isnull()]
```

```python
# Assuming Delhi to Cochin flight has one stop in between the two

df_final['Total_Stops'] =   df_final['Total_Stops'].fillna(1)
df_final['Total_Stops'].unique()
```

```python
# dropping the route column

df_final.drop('Route',axis=1,inplace=True)
df_final.head()
```

```python
import re

def dur_min(x):
    dur=None
    y=x.split(" ")
    if len(y)==2:
        if 'h' in y[0] and 'm' in y[1]:
            th = re.findall('\d+',y[0])
            for i in th:
                dur_h = int(i)*60

            tm = re.findall('\d+',y[1])
            for i in tm:
                dur_m = int(i)
            dur=dur_h+dur_m
    if len(y)==1:
        if 'h' in y[0]:
            th = re.findall('\d+',y[0])
            for i in th:
                dur_h = int(i)*60
            dur=dur_h
        else:
            tm = re.findall('\d+',y[0])
            for i in tm:
                dur_m = int(i)
            dur=dur_m    
    return dur
#     print(dur)
# dur_min('23h 45m')

df_final['Duration_min'] = df_final['Duration'].apply(dur_min)
df_final.head()
```

```python
# To derive hour from duration, same logic to be applied to get mins from duration

df_final['duration_hour']=df_final['Duration'].str.split(' ').str[0].str.split('h').str[0]
```
